```python
import pandas as pd
```

```python
raw = pd.read_csv('../data/cbb/tmp.csv')
raw
```

```python
x = list(raw.raw)
x

# make list with each year data as header
# make list ['1', '2', '3', '4'], loop through

seeds1 = [x[i] for i in range(len(x)) if x[i] == '1']
schools1 = [x[i+1] for i in range(len(x)) if x[i] == '1']
loc1 = [x[i+6].split('at ')[1] for i in range(len(x)) if x[i] == '1']
# print(loc1)

seeds2 = [x[i] for i in range(len(x)) if x[i] == '2']
schools2 = [x[i+1] for i in range(len(x)) if x[i] == '2']
loc2 = [x[i-1].split('at ')[1] for i in range(len(x)) if x[i] == '2']
# print(loc2)

seeds3 = [x[i] for i in range(len(x)) if x[i] == '3']
schools3 = [x[i+1] for i in range(len(x)) if x[i] == '3']
loc3 = [x[i+6].split('at ')[1] for i in range(len(x)) if x[i] == '3']
# print(loc3)

seeds4 = [x[i] for i in range(len(x)) if x[i] == '4']
schools4 = [x[i+1] for i in range(len(x)) if x[i] == '4']
loc4 = [x[i+6].split('at ')[1] for i in range(len(x)) if x[i] == '4']

seeds = seeds1+seeds2+seeds3+seeds4
schools = schools1+schools2+schools3+schools4
locs = loc1+loc2+loc3+loc4

obj = {'seed': seeds, 'school': schools, 'sites': locs}
df = pd.DataFrame(obj)
df

tmp = ['2019' + str(d) for d in list(df.index.values)]

df['id'] = tmp
df
```
